```python
from selenium import webdriver
from random import randint
import time
import json
```

```python
driver = webdriver.Chrome('C:/Users/Dell/Downloads/chromedriver_win32/chromedriver.exe')
```

```python
try:
    driver.set_window_size(randint(800, 1080), randint(1024, 1920))
    # get all links
#     links = []
    driver.get('https://www.geeksforgeeks.org/puzzles/')
    puzLink =  driver.find_elements_by_tag_name('ol')[0].find_elements_by_tag_name('a')[:-1]
    for link in puzLink:
        print(link.get_attribute("href"))
#         links.append(link.get_attribute("href"))
    
except Exception as e:
    print('Bakchodi ho gayi',str(e))
    
```

```python
links = [link.get_attribute("href") for link in puzLink]
links
```

```python
output = []
def scrapGFG(link):
    try:
        driver.get(link)
        time.sleep(2)
        name = driver.find_elements_by_class_name('title')[0].text 
        print(name)
        block = driver.find_elements_by_class_name('text')[-1]
        block = block.find_elements_by_tag_name('p')
        for i,p in enumerate(block):
            print(p.text,i)
        pText = [p.text for p in block]
        return name, pText
    except Exception as e:
        print('Bakchodi ho gayi',str(e))
for link in links:
    res = scrapGFG(link)
    output.append(res)
```

```python
final = []
for out in output:
    ques ={}
    ques['name'] = out[0]
    ques['question'] = out[1][0]
    ques['answer']= []
    for p in out[1]:
        if len(p.strip()) >5:
            ques['answer'].append(p.strip())
    ques['answer'] = (' ').join(ques['answer'])
    final.append(ques)
print(final)
```

```python
with open(f'result-{}.json', mode= 'w') as f:
    f.write(json.dumps(final))
#     f.append() and mode = 'a'
```
